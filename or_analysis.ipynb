```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import kagglehub
```

```python
path = kagglehub.dataset_download("nurielreuven/coffee-consumption-by-country-2022")

print("Path to dataset files:", path)
```

```python
os.listdir(path)
```

```python
csv_path = os.path.join(path, 'CoffeeConsumption.csv')
df = pd.read_csv(csv_path)
df.head()
```

```python
df.to_csv('CoffeeConsumption.csv', index=False)
```

```python
df.info()
```

```python
df_clean = df.dropna()
df_clean.head()
```

```python
top_countries = (
    df_clean.sort_values("perCapitaCons2016", ascending=False)
    .head(10)
)

import matplotlib.pyplot as plt

plt.figure()
plt.bar(top_countries["country"], top_countries["perCapitaCons2016"])
plt.xticks(rotation=45)
plt.ylabel("Coffee consumption per capita (2016)")
plt.title("Top 10 Coffee Consuming Countries per Capita")
plt.show()
```

This notebook analyzes a public coffee consumption dataset.
After basic data cleaning, the top 10 countries by coffee consumption per capita
were visualized using a bar chart.
